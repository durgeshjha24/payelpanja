```python
pivottranFF=pivottran.unstack()

pivottranFF.SALES_VALUE.fillna(0 ,inplace=True)
```

```python
# Joining of Demography data & transactional dataset

mergedata = pd.merge(Demodata,pivottranFF,on="household_key",how="left")
```
